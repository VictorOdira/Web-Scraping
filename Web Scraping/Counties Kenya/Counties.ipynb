```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.devolution.go.ke/county-information/"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    title = soup.title.text
    print(f"Title: {title}")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
table = soup.find(name = "table", class_ = "table table-hover")
table_rows = table.find_all(name = "tr")
data_list = []
for row in table_rows[1:]:
    columns = row.find_all(name = "td")
    data = [col.text.strip() for col in columns]
    #{
    #        "Code": columns[0].text.strip(),
    #        "County": columns[1].text.strip(),
    #        "Size (Km^2)": columns[2].text.strip(),
    #        "Population (NPHC 2009)": columns[3].text.strip(),
    #        "Capital": columns[4].text.strip(),
    #        "Governor": columns[5].text.strip(),
    #        "Deputy Governor": columns[6].text.strip(),
    #        "Women Representative": columns[7].text.strip(),
    #        "Main Economic Activities": columns[8].text.strip(),
    #    }
    data_list.append(data)

    # Print the extracted data
    for data in data_list:
        print(data)
```

```python
df = pd.DataFrame(data_list,
                  columns = ["Code", "County", "Size (Km^2)", "Population (NPHC 2009)",
                             "Capital", "Governor",  "Deputy Governor",  "Women Representative",
                            "Main Economic Activities"])
```
