```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.worldfootball.net/alltime_table/eng-premier-league/"
response = requests.get(url)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    title = soup.title.text
    print(f"Page Title:{title}")
else: 
    print(f"Failed, response status code is {response.status_code}")
```

```python
table = soup.find("table", class_="standard_tabelle")
print(table)
```

```python
rows = table.find_all("th")
print(rows)
```

```python
data = [row.text.strip() for row in rows]
epl_table = pd.DataFrame(columns = data)
epl_table
```

```python
column_data = table.find_all("tr")
for row in column_data[1:]:
    row_data = row.find_all("td")
    individual_row_data = [data.text.strip () for data in row_data]
    if '' in individual_row_data:
        individual_row_data.remove('')
    print(individual_row_data)
    length = len(epl_table)
    epl_table.loc[length] = individual_row_data
```

```python
epl_table
```

```python
epl_table = epl_table.rename(columns={
    '#': 'Position', 
    'M.': 'Played Matches',
    'W':'Won', 
    'D':'Draw', 
    'L':'Lost',
    'goals': 'Goals',
    'Dif':'Goal Difference',
    'Pt.':'Points'
})
```

```python
epl_table.to_csv("epl_table_of_all_time.csv")
```
